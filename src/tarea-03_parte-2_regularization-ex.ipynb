# Regularización

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
x = np.arange(0, 10, .5).reshape(20, 1)
y = x * np.sin(x)
plt.scatter(x, y)
plt.show()
```

# Instrucciones:

Completar el código para que:
- Calcule los vectores extendidos en "X_ext" usando PolynomialFeatures
- Entrenar Ridge(alpha=0.0, normalize=True) con X_ext y guardar las predicciones en "pred"

```python
orders = [2, 3, 4, 5, 6]
colors = ["r", "g", "y", "b", "m"]
plt.figure(figsize=(20,10))
plt.scatter(x, y)
for count, degree in enumerate(orders):
    trans = PolynomialFeatures(degree)
    X_ext = trans.fit_transform(x)
    clf = Ridge(alpha=0.0, normalize=True)
    clf.fit(X_ext, y)
    pred = clf.predict(X_ext)
    plt.plot(x, pred, c=colors[count], label="Grado: %d" %degree)
plt.legend(loc='lower left')
plt.show()
```

# Instrucciones:

- Elegir un grado de polinomio de la gráfica anterior que aproxime bien los datos.
- Extender los vectores de características a el grado elegido y guardar en X_ext.

Dentro del for
- Entrenar Ridge con regularización alpha=a
- Poner las predicciones en "pred"

```python
alpha = [0, .001, .001, .1, 0.5]
orden = 6 # poner order elegido
trans = PolynomialFeatures(orden)
X_ext = trans.fit_transform(x) # vectores extendidos
plt.figure(figsize=(20,10))
plt.scatter(x, y)
for i, a in enumerate(alpha):
    clf = Ridge(alpha=a, normalize=True)
    clf.fit(X_ext, y)
    pred = clf.predict(X_ext)
    plt.plot(x, pred, c=colors[i], label="alpha: %f" %a)
plt.legend(loc='lower left')
plt.show()
```
